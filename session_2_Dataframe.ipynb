## creating dataframe
###  What is a DataFrame?

A DataFrame is a 2-dimensional table in pandas (rows + columns).
It is the most commonly used data structure for data analysis.

```python
# using lists
student_data=[
    [100,80,10],
    [90,70,7],
    [120,100,141],
    [80,50,2]
]
pd.DataFrame(student_data,columns=['iq','marks','package'])
```

```python
## using dict
student_dict={
    'iq':[100,90,120,80,0,0],
    'marks':[80,70,100,50,0,0],
    'package':[10,7,14,2,0,0]
}
students=pd.DataFrame(student_dict)
students
```

## selecting cols from a Dataframe
| Method                | Description                      |
| --------------------- | -------------------------------- |
| `df["col"]`           | Select single column (Series)    |
| `df[["col1","col2"]]` | Select multiple columns          |
| `df.col`              | Dot notation (simple names only) |
| `df.iloc[:, index]`   | Select by column index           |
| `df.loc[:, label]`    | Select by column name            |
| `df.select_dtypes()`  | Select by data type              |

## selecting rows from a dataframe
| Method              | Description                   |
| ------------------- | ----------------------------- |
| `df.iloc[pos]`      | Select by index position      |
| `df.loc[label]`     | Select by index label         |
| `df[df["col"] > x]` | Conditional filtering         |
| `df.query()`        | Query using string expression |
| `df.head()`         | First N rows                  |
| `df.tail()`         | Last N rows                   |
| `df[start:end]`     | Row slicing                   |

## selecting both rows and cols
| Method                          | Meaning                          |
| ------------------------------- | -------------------------------- |
| `df.loc[row_label, col_label]`  | Select by **names/labels**       |
| `df.iloc[row_index, col_index]` | Select by **index numbers**      |
| `:`                             | Select **all**                   |
| `[ ]`                           | Select **specific** rows/columns |
| `start:end`                     | Select **range**                 |

## filtering Dataframe

| `between()`             | Check range           |                |

| `isnull()`              | Null values           |                |
