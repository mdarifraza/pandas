## üü¶ 1. **Concatenation (`pd.concat`)**

### **Definition**

`pd.concat` is used to **glue DataFrames together** either vertically (rows) or horizontally (columns).

### **Syntax**

```python
pd.concat([df1, df2], axis=0, ignore_index=False)
```

### **When to Use**

* When datasets have the **same columns** and you want to add more rows.
* When you want to add columns **without using keys**.

### **Types**

#### ‚úÖ Vertical Concat (add rows)

```python
pd.concat([df1, df2], axis=0)
```

#### ‚úÖ Horizontal Concat (add columns)

```python
pd.concat([df1, df2], axis=1)
```

#### ‚úÖ Reset index after concat

```python
pd.concat([df1, df2], ignore_index=True)
```

---

## üü¶ 2. **Merging (`pd.merge`)**

### **Definition**

`pd.merge` combines DataFrames using a **common column (key)** similar to SQL joins.

### **Syntax**

```python
pd.merge(df1, df2, on='key', how='inner')
```

### **When to Use**

* When you want to combine two tables based on a **matching column**.
* Similar to **SQL Join** operations.

### **Types of Merge**

#### 1Ô∏è‚É£ Inner Merge (default)

Returns rows that match in both tables.

```python
pd.merge(df1, df2, on='id', how='inner')
```

#### 2Ô∏è‚É£ Left Merge

Keeps all rows from left DataFrame.

```python
pd.merge(df1, df2, on='id', how='left')
```

#### 3Ô∏è‚É£ Right Merge

Keeps all rows from right DataFrame.

```python
pd.merge(df1, df2, on='id', how='right')
```

#### 4Ô∏è‚É£ Outer Merge

Returns all rows from both DataFrames.

```python
pd.merge(df1, df2, on='id', how='outer')
```

### **Merge on multiple columns**

```python
pd.merge(df1, df2, on=['id', 'date'])
```

---

## üü¶ 3. **Joining (`df.join`)**

### **Definition**

`df.join` is used to combine DataFrames based on **index**.

### **Syntax**

```python
df1.join(df2, how='left')
```

### **When to Use**

* When you want to join using **index instead of columns**.
* Shortcut for merging on index.

### **Types of Join**

Same as merge:

* `left`
* `right`
* `inner`
* `outer`

### **Example**

```python
df1.join(df2, lsuffix='_A', rsuffix='_B')
```

Adds suffixes when columns overlap.

---

## üü© Summary Table

| Operation       | Function    | Based On            | Best Use Case                    |
| --------------- | ----------- | ------------------- | -------------------------------- |
| **Concatenate** | `pd.concat` | Axis (rows/columns) | Stacking datasets (same columns) |
| **Merge**       | `pd.merge`  | Key column          | SQL-like joins                   |
| **Join**        | `df.join`   | Index               | Index-based merging              |

---

## üü£ Interview Tips

* Use **concat** for adding rows (like appending monthly data).
* Use **merge** when combining tables using keys.
* Use **join** when your keys are **index-based**.
* Merging is the most versatile and preferred for relational datasets.
