## pandas
##  What is Pandas?

**Pandas** is a **Python library** used for **data analysis and data manipulation**.  
It allows you to easily handle **structured data** (like tables, spreadsheets, CSV files, etc.) using powerful and flexible data structures.

---

###  Key Points:
- Pandas stands for **"Python Data Analysis Library"**.  
- It is built on top of **NumPy**.  
- Used widely in **data science, machine learning, and analytics**.  
- It helps in:
  - Cleaning data  
  - Transforming data  
  - Analyzing data  
  - Handling missing values  
  - Working with time series data  

---

###  Main Data Structures:
| Data Structure | Description | Example |
|----------------|--------------|----------|
| **Series** | 1D labeled array (like a column in Excel) | `[10, 20, 30]` |
| **DataFrame** | 2D labeled data (like an Excel sheet or SQL table) | `rows  columns` |

---

###  Example:
```python

# Create a simple DataFrame

    'Name': ['Arif', 'Ravi', 'Neha'],
    'Age': [25, 28, 24]

## pandas series 
##  Pandas Series

A **Series** in Pandas is a **one-dimensional labeled array** that can hold any type of data  integers, strings, floats, or even objects.

You can think of a Series as a **single column** in an Excel sheet or a **list with labels (index)**.

---

###  Features of Series:
- It is **one-dimensional** (like a single column).
- Has **index labels** to access data easily.
- Can hold **different data types** (int, float, string, etc.).
- Built on **NumPy arrays**, so it supports vectorized operations.

## series attribute
##  Pandas Series Attributes

A **Series** object in Pandas has several **attributes** that give information about its data, type, size, and index.

These attributes **do not require parentheses `()`** because they are **not functions**, only **properties** of the Series.

---

###  Commonly Used Attributes

| Attribute | Description | Example | Output |
|------------|--------------|----------|---------|
| `s.values` | Returns the data as a **NumPy array** | `s.values` | `[10 20 30 40]` |
| `s.index` | Returns the **index (labels)** of the Series | `s.index` | `RangeIndex(start=0, stop=4, step=1)` |
| `s.dtype` | Returns the **data type** of the Series | `s.dtype` | `int64` |
| `s.shape` | Returns the **shape** (size in a tuple) | `s.shape` | `(4,)` |
| `s.size` | Returns the **number of elements** | `s.size` | `4` |
| `s.ndim` | Returns the **number of dimensions** (for Series always 1) | `s.ndim` | `1` |
| `s.name` | Returns or sets the **name of the Series** | `s.name = 'Marks'` | `'Marks'` |
| `s.is_unique` | Checks if all values are **unique** | `s.is_unique` | `True / False` |
| `s.hasnans` | Checks if Series has any **missing (NaN)** values | `s.hasnans` | `False` |
| `s.empty` | Checks if Series is **empty** | `s.empty` | `False` |

---

###  Example

```python
import pandas as pd

s = pd.Series([10, 20, 30, 40])

print("Values:", s.values)
print("Index:", s.index)
print("Data Type:", s.dtype)
print("Shape:", s.shape)
print("Size:", s.size)
print("Dimensions:", s.ndim)
print("Is Unique:", s.is_unique)
print("Has NaN:", s.hasnans)
print("Is Empty:", s.empty)

---

###  Remember:

- You can then **select a single column** from that DataFrame to make a **Series**.  



-    |  Task                      |  Code                                                       |
- | ---------------------------- | ------------------------------------------------------------- |

- | **Set index column**         | `pd.read_csv("file.csv", index_col="col")["target_col"]`      |

##  Pandas Series Indexing

Indexing in **Pandas Series** helps you **access, slice, or modify** elements.  
Each element in a Series has an **index** (like labels in a list).

---

###  Types of Indexing

|  Type |  Description |  Example |
|----------|----------------|------------|
| **Positional Indexing** | Access by integer position (like lists) | `s[0]`, `s[1:4]` |
| **Label-based Indexing** | Access by index labels | `s['a']`, `s['b':'d']` |
| **Boolean Indexing** | Access using condition | `s[s > 50]` |
| **Fancy Indexing** | Access multiple elements at once | `s[[0, 2, 4]]` or `s[['a', 'c']]` |

---

###  Example

```python
import pandas as pd

s = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])
print("Series:\n", s)

# Positional Indexing
print("\nPositional (s[0]):", s[0])
print("Slice by position (s[1:4]):\n", s[1:4])

# Label-based Indexing
print("\nLabel (s['c']):", s['c'])
print("Slice by label (s['b':'d']):\n", s['b':'d'])

# Boolean Indexing
print("\nValues > 30:\n", s[s > 30])

# Fancy Indexing
print("\nMultiple labels (s[['a', 'e']]):\n", s[['a', 'e']])

## copy and views
##  Copy and Views in Pandas Series

When you work with a **Pandas Series**, sometimes you create a **view** (reference)  
and sometimes a **copy** (independent duplicate).  
Understanding the difference helps prevent **unexpected changes** in your data.
|  Type  |  Meaning                 |  Effect of Change   |  Example          |
| -------- | -------------------------- | --------------------- | ------------------- |
| **View** | Reference to original data | Affects both          | `view_s = s`        |
| **Copy** | Separate object            | No effect on original | `copy_s = s.copy()` |

## boolean indexing series
##  Boolean Indexing in Pandas Series

**Boolean Indexing** means filtering data based on **True/False conditions**.  
It allows you to select values that satisfy a certain condition.

---

|  Task                  |  Example               |            |
| ------------------------ | ------------------------ | ---------- |
| Greater than 30          | `s[s > 30]`              |            |
| Less than or equal to 40 | `s[s <= 40]`             |            |
| Not equal to 20          | `s[s != 20]`             |            |
| Between two values       | `s[(s > 10) & (s < 50)]` |            |
| Using OR condition       | `s[(s < 20)              | (s > 40)]` |
| Negation (NOT)           | `s[~(s > 30)]`           |            |
