`value_counts()` is used to count how many times each **unique value** appears in a **Series** or a **DataFrame (with multiple columns)**.

---

##  When to Use `value_counts()`?

* To analyze distribution of values
* To count unique combinations (multi-column)

##  `sort_values()` (Series & DataFrame)

###  **Simple Definition:**

**`sort_values()` is used to arrange (sort) your data in increasing or decreasing order based on one or more columns.**

---

| Category      | Parameter / Usage                                            | Description                                 |
| ------------- | ------------------------------------------------------------ | ------------------------------------------- |
| Series        | `series.sort_values()`                                       | Sort a single column (Series) by its values |
| DataFrame     | `df.sort_values(by='col')`                                   | Sort DataFrame by one column                |
| Multiple Cols | `df.sort_values(by=['col1','col2'])`                         | Sort using multiple columns                 |
| ascending     | `ascending=True / False`                                     | True  lowhigh, False  highlow           |
| multi-asc     | `ascending=[True, False]`                                    | Different order for each column             |
| na_position   | `na_position='last'` (default)                               | Keep NaN at end                             |
| na_position   | `na_position='first'`                                        | Keep NaN at beginning                       |
| inplace       | `inplace=True`                                               | Sort the original DataFrame                 |
| inplace       | `inplace=False` (default)                                    | Returns a new sorted DataFrame              |
| Example       | `df.sort_values(by='Age', ascending=False)`                  | Sort by Age (descending)                    |
| Example       | `df.sort_values(by=['City','Age'], ascending=[True, False])` | Sort by multiple columns                    |

---

##  `rank()` (Series)

###  **Simple Definition:**

**`rank()` gives a ranking number to each value in a Series based on its order (small  big or big  small).**

---

###  **Syntax**

```python
series.rank()
```

---

###  **Important Parameters**

| Parameter / Usage            | Description                             |
| ---------------------------- | --------------------------------------- |
| `series.rank()`              | Default ranking (average rank for ties) |
| `method='average'` (default) | Same values get the **average** rank    |
| `method='min'`               | Same values get the **minimum** rank    |
| `method='max'`               | Same values get the **maximum** rank    |
| `method='first'`             | Ranks based on order of appearance      |
| `ascending=True`             | Small  big (default)                   |
| `ascending=False`            | Big  small                             |
| `na_option='keep'` (default) | Keep NaN in rank                        |
| `na_option='top'`            | NaN considered smallest                 |
| `na_option='bottom'`         | NaN considered largest                  |
| Example                      | `s.rank(ascending=False)`               |
| Example                      | `s.rank(method='min')`                  |

---

# ## dropna()  Series + DataFrame

`dropna()` is used to **remove missing values (NaN)** from a Series or DataFrame.

---

#  **1. dropna() on Series**

###  Removes NaN values and returns a new Series

```python
s.dropna()
```
#  **2. dropna() on DataFrame**

###  Basic usage

```python
df.dropna()
```

Removes **rows** that contain **any NaN**.

---

#  Important Parameters

## 1 **axis**

* `axis=0`  drop rows (default)
* `axis=1`  drop columns

```python
df.dropna(axis=1)      # remove columns with NaN
```

---

## 2 **how**

* `how='any'`  drop if **any** NaN present (default)
* `how='all'`  drop only if **all** values are NaN

```python
df.dropna(how='all')
```

---

## 3 **thresh**

Keep rows/columns with **at least N non-null values**.

```python
df.dropna(thresh=2)
```

---

## 4 **subset**

Check NaN only in selected columns.

```python
df.dropna(subset=['Age', 'Salary'])
```

---

## 5 **inplace**

Modify the original DataFrame.

```python
df.dropna(inplace=True)
```

---

#  Quick Summary Table

| Parameter | Meaning                    |
| --------- | -------------------------- |
| axis      | 0 = rows, 1 = columns      |
| how       | 'any' or 'all'             |
| thresh    | Min non-null values needed |
| subset    | Columns to check           |
| inplace   | Modify original            |

# ## apply()  Series + DataFrame

`apply()` is used to **apply a function** on each element, row, or column of a Series or DataFrame.

---

#  1. apply() on **Series**

Runs a function on **each value** of a Series.

###  Syntax

```python
s.apply(function)
```
#  2. apply() on **DataFrame**

It can apply a function **column-wise** or **row-wise**.

###  Syntax

```python
df.apply(function, axis=0/1)
```

###  axis Meaning

| axis           | applies to  | meaning  |
| -------------- | ----------- | -------- |
| 0 or 'index'   | each column | default  |
| 1 or 'columns' | each row    | row-wise |

---


#  When to use apply?

* Apply logic on each value
* Create new transformed column
* Perform row-wise calculations
* Custom operations not available in built-in Pandas functions

---

#  Quick Summary Table

| Task                     | Code                                                       |
| ------------------------ | ---------------------------------------------------------- |
| Apply function on Series | `s.apply(func)`                                            |
| Column-wise apply        | `df.apply(func)`                                           |
| Row-wise apply           | `df.apply(func, axis=1)`                                   |
| Lambda inside apply      | `df['A'].apply(lambda x: x*2)`                             |
| Conditional apply        | `df['Age'].apply(lambda x: "Adult" if x>=18 else "Minor")` |

---

````python
points_df=pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]

    }
)
points_df
````

````python
def euclidian(row):
    pt_A=row['1st point']
    pt_b=row['2nd point']

    return ((pt_A[0]-pt_b[0])**2 +(pt_A[1]-pt_b[1])**2)**0.5
````

````python
points_df['distance']=points_df.apply(euclidian,axis=1)
points_df
````
